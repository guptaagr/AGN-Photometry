```python
filters = ['uw1', 'uw2', 'um2', 'uuu']
radii = [5.000, 7.000, 10.000, 15.000, 20.000, 25.000, 30.000, 35.000]
for filterr in filters:
    mags = []
    for radius in radii:
        if os.path.exists(f"maghist_{filterr}_{str(radius)}00_00096108001.fits"):
            maghist = fits.open(f"maghist_{filterr}_{str(radius)}00_00096108001.fits")
            mag_data = Table(maghist[1].data)
            mags.append(mag_data['CORR_RATE'].mean())
        else:
            continue
    if mags:
        plt.plot(radii, mags, label=f"00096108001_{filterr}")
        plt.legend();
```
